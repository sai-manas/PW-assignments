**Q5. How are DataFrame.size() and DataFrame.shape() different?**


In Pandas, DataFrame.size and DataFrame.shape are two different attributes that provide distinct information about a DataFrame:

**DataFrame.size:**

- DataFrame.size returns the total number of elements in the DataFrame, which is equal to the product of the number of rows and the number of columns.
- It returns a single integer value representing the total size of the DataFrame, including all cells or elements.
- You can access it as an attribute, i.e., df.size where df is your DataFrame.

**DataFrame.shape:**

- DataFrame.shape returns a tuple representing the dimensions of the DataFrame. It provides the number of rows and columns in the DataFrame.
- It returns a tuple with two elements: the first element is the number of rows, and the second element is the number of columns.
- You can access it as an attribute, i.e., df.shape where df is your DataFrame.

```python
import pandas as pd

data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}

df = pd.DataFrame(data)

# Using DataFrame.size to get the total number of elements
size = df.size  # This will be 6 (3 rows * 2 columns)

# Using DataFrame.shape to get the dimensions (rows, columns)
shape = df.shape  # This will be (3, 2) indicating 3 rows and 2 columns

print("DataFrame size:", size)
print("DataFrame shape:", shape)
```

**Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.**

**The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.**

```python
data1 = {'Email': ['john.doe@gmail.com', 'jane.smith@yahoo.com', 'bob@google.com','manas@pw.com']}
df_mail = pd.DataFrame(data1)
df_mail
```

```python
def extract_username(df):
    df['Username'] = df['Email'].str.split('@').str[0]
    return df

extract_username(df_mail)
```

**Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.**

```python
date_time = pd.date_range(start='2023-09-05',end='2023-09-14')
date_time
```

```python
date_df = pd.DataFrame({"date":date_time,"sales":[100,200,300,400,500,600,700,800,900,1000]})
date_df
```

```python
def moving_average(df):
    df['MovingAverage'] = df['sales'].rolling(window=7).mean()
    return df

moving_average(date_df)
```
