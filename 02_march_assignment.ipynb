**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.**


Matplotlib is a widely-used Python library for creating static, animated, and interactive data visualizations. It provides a wide range of tools for creating various types of plots and charts, making it a valuable tool for data analysis, scientific research, and data visualization tasks. Matplotlib is highly customizable and can be used to create publication-quality plots.

Here's why Matplotlib is used:

- Data Visualization: Matplotlib is used to visualize data by creating a wide variety of plots and charts, helping users understand and interpret data more effectively.

- Customization: It offers extensive customization options, allowing users to control every aspect of the plot's appearance, including colors, markers, labels, fonts, and more.

- Publication-Quality Plots: Matplotlib enables the creation of publication-ready plots and figures, suitable for use in research papers, reports, presentations, and publications.

- Integration: It can be integrated with other libraries and tools commonly used in the data science ecosystem, such as NumPy, Pandas, and Jupyter notebooks.

- Interactive and Animated Visualizations: Matplotlib supports interactive plotting and animation, making it suitable for creating dynamic and interactive visualizations for presentations and web applications.

Here are five types of plots that can be created using the Pyplot module of Matplotlib:

- Line Plot: Line plots are used to display data points connected by straight lines. They are often used to visualize trends or changes in data over time.


- Scatter Plot: Scatter plots display individual data points as markers on a two-dimensional plane. They are useful for visualizing the relationship between two variables.


- Bar Plot: Bar plots or bar charts represent data with rectangular bars, where the length of each bar corresponds to the value of a variable. They are often used to compare values across categories.


- Histogram: Histograms are used to visualize the distribution of a continuous dataset. They group data into bins and display the frequency or probability of data falling into each bin.


- Pie Chart: Pie charts represent data in a circular graph, where each wedge (slice) represents a category's proportion in the whole. They are useful for showing the composition of a whole.



**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.**

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("SCATTER PLOT")
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:**

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


- The subplot() function in Matplotlib is used to create multiple plots or subplots within a single figure. It allows you to arrange multiple plots in a grid-like fashion, specifying the number of rows and columns. This is useful when you want to compare or display multiple sets of data side by side for better visualization.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("Plot 4")

plt.tight_layout()
plt.show()
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])




A bar plot, also known as a bar chart or bar graph, is a data visualization that uses rectangular bars to represent data values. Each bar's length or height is proportional to the value it represents. Bar plots are commonly used to compare and display categorical data, making them effective for visualizing data distributions, comparisons, and trends.

Bar plots are used for several purposes:

- Comparisons: Bar plots are excellent for comparing values across different categories or groups. They make it easy to see which categories have higher or lower values.

- Distribution: Bar plots can show the distribution of data within different categories. For example, you can use them to display the frequency or count of items in each category.

- Trends: Over time, bar plots can be used to visualize trends and changes in data values, particularly when comparing data points at different time periods or conditions.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(8,4))
plt.bar(company,profit)
plt.title("Bar Graph")
plt.xlabel("company")
plt.ylabel("profit")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(company,profit)
plt.title("Bar Graph")
plt.xlabel("profit")
plt.ylabel("company")
plt.show()
```

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)


A box plot, also known as a box-and-whisker plot, is a data visualization tool used to display the distribution and spread of a dataset. It provides a summary of key statistical measures such as the median, quartiles, and potential outliers. Box plots are useful for:

- Identifying Central Tendency: Box plots show the median (or middle) value of the dataset, which represents the central tendency.

- Visualizing Spread: They display the range between the first quartile (Q1) and the third quartile (Q3), known as the interquartile range (IQR), which indicates the spread of the data.

- Detecting Outliers: Box plots can reveal potential outliers in the data, which are data points that are significantly different from the rest of the dataset.

- Comparing Distributions: Box plots allow you to compare the distributions of multiple datasets or groups side by side.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize=(8,4))
plt.boxplot([box1,box2],labels=['Box 1', 'Box 2'])
plt.ylabel("values")
plt.title('Box Plot of Box1 and Box2')
plt.show()
```
