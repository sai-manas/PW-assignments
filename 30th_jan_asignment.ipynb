**1. Write a program to accept percentage from the user and display the grade according to the following**


```python
n = int(input("What is your percentage:"))
if n in range(1,101):
    if n > 90 :
        print(" Your have obtained grade\"A\"")
    elif n>80 and n<=90:
        print(" Your have obtained grade \"B\"")
    elif n>=60 and n<=80:
        print(" Your have obtained grade \"C\"")
    else:
        print(" Your have obtained grade \"D\"")
else:
    print("Wrong input('This code only takes 1 to 100 percentages')")
    
```

**2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria:**


```python
i = int(input("Enter the price of the bike: "))
if i >= 0:
    if i > 100000:
        print("15% Road tax to be paid for price of {} which is: {}".format(i,i*0.15))
    elif i >50000 and i <= 100000:
        print("10% Road tax to be paid for price of {} which is: {}".format(i,i*0.1))
    else:
        print("5% Road tax to be paid for price of {} which is: {}".format(i,i*0.05))
else:
    print("Enter valid price")
```

**3. Accept any city from the user and display monuments of that city.** 


```python
c = str(input("Enter the name of the city: "))
if c.lower() == "Delhi".lower():
    print("'Red Fort' is the monument of ",c.upper())
elif c.lower() == "Agra".lower():
    print("'Taj Mahal' is the monument of ",c.upper())
elif c.lower() == "Jaipur".lower():
    print("'Jal Mahal' is the monument of ",c.upper())
else:
    print("NA")
```

**4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.**

```python
def rep_check(n):
    count = 0
    while n > 10:
        n /= 3
        count += 1
    return count
```

```python
num = int(input("Enter the number: "))
print(f"{num} can be divied by 3 for {rep_check(num)} times before it is less than or equal to 10")
```

**6. Use nested while loop to print 3 different pattern.**

```python
n = 1
while n<=5:
    i = 1
    while i<=n:
        print("*",end ="")
        i+=1
    print()
    n+=1
```

```python
n = 5
while n>=1:
    i = 1
    while i<=n:
        print("*",end ="")
        i+=1
    print()
    n-=1
```

```python
def pattern(n):
    i = 0
    while i<=n:
        print(" "*(n-i)+"*"*((2*i)+1),end ="")
        i+=1
        print()
```

```python
pattern(10)
```
